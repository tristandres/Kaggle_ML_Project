```python
import pandas as pd
import numpy as np
import seaborn as sns
import datetime as dt
import random as rn
import matplotlib.pyplot as plt
import sklearn
from sklearn.feature_selection import chi2
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder
```

```python
data = pd.read_csv('Train_Kaggle_Data_11_12PM.csv')
data['MSSubClass'] = data['MSSubClass'].astype(str)
data['OverallQual'] = data['OverallQual'].astype(str)
data['OverallCond'] = data['OverallCond'].astype(str)
```

```python
all_cat = list(data.loc[:, data.dtypes == object].columns)
```

```python
#to be able to use the Chi Squared test on two categorical variables,
#they need to be transformed into integers
data1 = data.copy()
le = preprocessing.LabelEncoder()
for i in range(len(all_cat)):
    le.fit(data[all_cat[i]])
    data1[all_cat[i]] = le.transform(data[all_cat[i]])
```

```python
#H0 features ARE independent (no relationship)
#Ha features related (there is a relationship)
#if p < 0.05, reject H0, features are related. 

pairs_keep = []
drop = []

for cat_base in range(len(all_cat)):
    rest_of_cats = all_cat[cat_base + 1:]
    X = np.array(data1[all_cat[cat_base]])
    
    for cat in range(len(rest_of_cats)):
        
        y = np.array(data1[rest_of_cats[cat]])
        
        if sklearn.feature_selection.chi2(X.reshape(-1,1),y)[1] < 0.01:
            #if Rsq for x vs price > Rsq y vs price:
                #add y to list to drop
            pairs_keep.append(all_cat[cat_base])
            pairs_keep.append(rest_of_cats[cat])
            


pairs_total= len(all_cat)*(len(all_cat)+1)/2 #1081
len(pairs_keep)
```

```python
#for i in set(keep_pairs):
L1 = list(set(pairs_keep))
df = pd.DataFrame(L1,columns= ['Category'])
L2 = []
for i in L1:
    L2.append(pairs_keep.count(i))
df['%_related'] = L2
df['%_related'] = round(df['%_related']/46 * 100,1)
df1 = df.sort_values(by='%_related',ascending=False)
#L3 = list(range(1,46))
#df1['rank'] = L3
df1 #number times a category was shown to be related to another category
```

```python
cats_ = all_cat[1:]
cats_
L4 = []
L5 = []
for i in cats_:
    X = np.array(data1['MSSubClass'])
    y = np.array(data1[i])
    if sklearn.feature_selection.chi2(X.reshape(-1,1),y)[1] > 0.01:
        L4.append(i)
    
for i in range(len(L4)):
    L6 = L4[i+1:]
    for n in range(len(L6)):
        X = np.array(data1[L4[i]])
        y = np.array(data1[L6[n]])
        if sklearn.feature_selection.chi2(X.reshape(-1,1),y)[1] > 0.01:
            L5.append(L4[i])
            L5.append(L6[n])
chi_sq_cats = list(set(L5))
chi_sq_cats = list(['MSSubClass']) + chi_sq_cats[:-1]
chi_sq_cats
```
